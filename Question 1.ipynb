```python
import numpy as np
np.set_printoptions(threshold=np.inf)
import matplotlib.pyplot as plt
```

Generate Data using given vectors

```python
AV = [0, 20, 0, 0, 0, 0]
IV = [30, 45, 60, 40, 40, 40]
DO = [15, 20, 25, 15, 20, 25]
```

```python
def createTC(av, iv, do):
    TCn = np.zeros(240, dtype=int)
    inc = 240/iv
    while inc >= 1:
        for j in range(do):
            TCn[av+j] = 1
        av = av + iv
        inc -= 1
    
    return TCn
s = (240, 6)
TCL = []
for i in range(6):
    av = AV[i]
    iv = IV[i]
    do = DO[i]
    inc = 240/iv
    TCL.append(createTC(av, iv, do))

TC = np.asmatrix(TCL)
print(TC)    
```

```python
# Normalising the generated data
TCLNormalised = []
for i in range(6):
    mu = np.mean(TCL[i])
    std = np.std(TCL[i])
    TCLNormalised.append( (TCL[i]-mu)/std)
TCNormalised = np.asmatrix(TCLNormalised)
```

```python
# plotting the graphs
x = range(240)
fig = plt.figure()

fig.set_figheight(10)
fig.set_figwidth(20)

for i in range(6):
    ax = fig.add_subplot(231+i) 
    ax.plot(x, TCLNormalised[i])
plt.show()
```
