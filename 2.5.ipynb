```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# determine important variables
N = 240
V = 441
x1 = 21
x2 = 21
L = 6
nsrcs = 6

SMcmap = sns.cubehelix_palette(start=.5, rot=-.75, as_cmap=True)
TCcmap = sns.cubehelix_palette(start=.5, rot=-.5, as_cmap=True)
TCcolor = "darkturquoise"
SMcolor = "palegreen"

# import generated dataset
APCR = pd.read_pickle("data/APCR.pkl").to_numpy()
DPCR = pd.read_pickle("data/DPCR.pkl").to_numpy()
```

```python
x = range(N)
fig, axes = plt.subplots(6, 2) 
fig.set_figheight(30)
fig.set_figwidth(20)

for i in range(nsrcs):
    axes[i][0].set_title(f"Retrieved Spatial Maps{i+1} from PCR")
    sns.heatmap(APCR[i].reshape(x1,x2), cmap=SMcmap, ax = axes[i,0])
    axes[i][1].set_title(f"Retrieved Time Courses{i+1} from PCR")
    axes[i][1].plot(x, DPCR[:, i], color = TCcolor)

plt.savefig('plots/PCR maps.png')
plt.show()
```
